## matplotlib example workflow

```python
# step1 import matplotlib

%matplotlib inline
import matplotlib.pyplot as plt

# step2 prepare data

x = [1,2,3,4]
y = [1,4,9,16]

#step3 setup figure and axes

fig,ax = plt.subplots(figsize=(5,5))  # width and height (10,10)

#step 4 plot data

ax.plot(x,y)

#step 5 customize

ax.set(title="simple plot",xlabel="x_axis",ylabel="y-axis")

#step6 save plot

fig.savefig("images/simple_plot.png")
```

# make a plot from dictinary

```python
my_food_price = {"pizza":120,"briyani":150,"chicken messala":180,"dossa":50}

fig,ax = plt.subplots()
ax.bar(my_food_price.keys(),my_food_price.values())
ax.set(title="my best food",xlabel="food type",ylabel="price in ruppi");
```

```python
# horizontal bar
fig,ax = plt.subplots()
ax.barh(list(my_food_price.keys()),list(my_food_price.values()));
```

```python
# histogram or hist

x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
ax.set(title="normal distribution");
```

## option of plotting subplots

```python
#option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                        ncols=2,
                                         figsize=(10,10))
ax1.plot(x,x/2);
ax1.set(title="figure1");
ax2.scatter(np.random.random(10),np.random.random(10));
ax2.set(title="figure2");
ax3.bar(my_food_price.keys(),my_food_price.values());
ax3.set(title="figure3");
ax4.hist(x);
ax4.set(title="figure4");
```

```python
# option 2
# using like a list []
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,10))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(my_food_price.keys(),my_food_price.values());
ax[1,1].hist(x);
```
